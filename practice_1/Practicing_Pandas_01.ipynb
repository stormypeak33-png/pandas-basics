```python
hours = np.array([1, 2, 3, 4, 5, 6, 2, 4, 3, 5])
result = np.array([0, 0, 0, 1, 1, 1, 0, 1, 0, 1])

df = pd.DataFrame({"Hours": hours, "Result": result})             #Making it as a DataFrame
df["Result_label"] = df["Result"].map({0:"Fail", 1:"Pass"})       #Adding Result Label

print(df.to_string(index = False))                                #Printing without index

avg_pass = hours[result == 1].mean()
avg_fail = hours[result == 0].mean()                              #Calculating mean

print(f"Average Pass : {avg_pass}")
print(f"Average Fail : {avg_fail}")                               #Printing results
```

```python
names = np.array(["Alice","Bob","Charlie","Diana","Eve","Frank","Grace","Hank","Ivy","Jake"])
hours = np.array([1, 2, 3, 4, 5, 6, 2, 4, 3, 5])
scores = np.array([40, 45, 50, 65, 70, 85, 42, 68, 55, 72])
result = np.array([0,  0,  0,  1,  1,  1,  0,  1,  0,  1])
subject = np.array(['Math', 'Science', 'Math', 'Science', 'Math', 'Science', 'Math', 'Science', 'Math', 'Science'])

#Making it as DataFrame
df = pd.DataFrame({
    "Names": names,
    "Hours": hours,
    "Scores" : scores,
    "Result": result,
    "Subject": subject
})

#Add Grade
df["Grade"] = np.where(df["Scores"] >= 60, "A", "B")

#Print Without Index
print(df.to_string(index = False))


print(df[ (df["Result"] == 1) & (df["Grade"] == "A")].to_string(index = False))
print()
print(df["Result"].map({
    0: "Fail",
    1: "Pass"
}).value_counts().to_string(index = False))
passed = df [df["Result"]==1]
print("=====Maximum Scores=====")
print(passed.loc[ passed["Scores"].idxmax() ])
print("=====Minimum score=====")
print(passed.loc[passed["Scores"].idxmin()])

failed = df[df["Result"] == 0 ]
print("Studied most of the hours still fails")
print(failed.loc[failed["Hours"].idxmax()])

total = len(df)
passed_count = len(df[df["Result"] == 1])
pass_rate = (passed_count/ total) * 100
print("Pass rate: ",pass_rate)

```

```python
import pandas as pd

# Sample data
data = {
    'Hours': [5, 3, 7, 2, 8, 4],
    'Score': [85, 62, 90, 45, 92, 78],
    'Result': ['Pass', 'Fail', 'Pass', 'Fail', 'Pass', 'Pass']
}
df = pd.DataFrame(data)

# Step 1: Add Subject column
df['Subject'] = ['Math', 'Math', 'Science', 'Science', 'Math', 'Science']

# Step 2: Group by Subject -> avg hours & avg score
subject_gb = df.groupby('Subject').agg(
    Avg_Hours=('Hours', 'mean'),
    Avg_Score=('Score', 'mean')
).reset_index()

# Step 3: Group by Result -> count, avg hours, avg score
result_gb = df.groupby('Result').agg(
    Count=('Result', 'count'),
    Avg_Hours=('Hours', 'mean'),
    Avg_Score=('Score', 'mean')
).reset_index()

# Step 4: Which subject has higher pass rate?
# Calculate pass rate per subject (True/False mean evaluates to the percentage of True)
df['IsPass'] = df['Result'] == 'Pass'
pass_rates = df.groupby('Subject')['IsPass'].mean()
best_subject = pass_rates.idxmax()

# Output Results
print("--- Subject Group --- \n", subject_gb)
print("\n--- Result Group --- \n", result_gb)
print(f"\nSubject with highest pass rate: {best_subject} ({pass_rates[best_subject]*100}%)")
```

```python
import numpy as np
import pandas as pd

drivers = np.array(["FedEx", "UPS", "FedEx", "DHL", "UPS", "DHL", "FedEx", "DHL", "UPS", "FedEx"])
distance = np.array([12, 25, 8, 45, 18, 30, 15, 50, 22, 10])     # in miles
fees = np.array([15, 30, 12, 65, 22, 45, 18, 70, 28, 14])        # in dollars
status = np.array([1, 1, 0, 1, 0, 1, 1, 0, 1, 1])               # 1 = On-Time, 0 = Late

df = pd.DataFrame({
    "Drivers": drivers,
    "Distances": distance,
    "Fees": fees,
    "Status": status
})
df["Region"] = ['North', 'South', 'North', 'South', 'North', 'South', 'North', 'South', 'North', 'South']
avg = df.groupby('Region').agg(
    Avg_dist = ("Distances", "mean"),
    Avg_fee = ("Fees", "mean")
).reset_index()
print(avg)

dri = df.groupby("Drivers").agg(
    Count = ("Status", "count"),
    max_dist = ("Distances", "max"),
    avg_fees = ("Fees", "mean")
).reset_index()
print(dri)

on_time = df.groupby("Region").agg(
    Driver = ("Status", "mean")
).reset_index()
print(on_time.to_string(index = False))
```

```python
import numpy as np
import pandas as pd

titles = np.array(["Zelda", "Mario", "Halo", "Fifa", "Cod", "Pokemon", "Metroid", "Sonic", "Minecraft", "GTA"])
platforms = np.array(["Switch", "Switch", "Xbox", "PlayStation", "Xbox", "Switch", "Switch", "PlayStation", "Xbox", "PlayStation"])
prices = np.array([60, 60, 70, 70, 70, 60, 50, 40, 30, 70])
bestseller = np.array([1, 1, 0, 1, 0, 1, 0, 0, 1, 1]) # 1 = Bestseller, 0 = Regular seller

df = pd.DataFrame({
    "Titles": titles,
    "Platforms": platforms,
    "Prices": prices,
    "Bestsellers": bestseller
})
df["Genre"] = ['Action', 'RPG', 'Action', 'RPG', 'Action', 'RPG', 'Action', 'RPG', 'Action', 'RPG']
print(df.to_string(index = False))
group_by_genre = df.groupby("Genre").agg(
    Avg_price = ("Prices", "mean")
).reset_index()
print()
print(group_by_genre)

group_platform = df.groupby("Platforms").agg(
    count = ("Titles", "count"),
    Best = ("Bestsellers", "mean")
).reset_index()
print()
print(group_platform.to_string(index = False))
```

```python
dirty_data = {
    "Name":   ["Alice", "Bob", None, "Diana", "Eve", "Frank", None, "Hank", "Ivy", "Jake"],
    "Hours":  [1, 2, 3, None, 5, 6, 2, 4, None, 5],
    "Score":  [40, 45, 50, 65, None, 85, 42, 68, 55, None],
    "Result": [0, 0, 0, 1, 1, 1, 0, 1, 0, 1]
}
df = pd.DataFrame(dirty_data)
#Detecting Nulls
print(df.isnull())
print(df.isnull().sum())
print(df.isnull().sum().sum())
#Filling nulls with zero
df["Score"] = df["Score"].fillna(0)

#Filling nulls with mean
df["Hours"] = df["Hours"].fillna(df["Hours"].mean())

#Filling with median / mode                                # Mode is -> Most Frequently occuring value
df["Score"] = df["Score"].fillna(df["Score"].median())
df["Name"] = df["Name"].fillna(df["Name"].mode()[0])       #Mode always returns a series [0, 1, 2, 3....] so [0] needs to extract it

```

```python
#Dropping nulls
df = df.dropna(subset = ["Name"])
print(df)
```

```python
import pandas as pd
import numpy as np

hr_data = {
    "EmpID":      [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    "Name":       ["Ravi", "Sara", None, "John", "Priya", "Tom", None, "Nina", "Arjun", "Lisa"],
    "Department": ["IT", "HR", "IT", "Finance", "IT", "HR", "Finance", "IT", "Finance", None],
    "Salary":     [50000, 60000, 55000, None, 70000, 48000, 62000, None, 75000, 53000],
    "Experience": [2, 5, 3, 7, 4, 1, 6, 3, 8, 2],
    "Rating":     [4, 5, 3, 4, 5, 2, 4, 3, 5, None],
    "Left":       [0, 0, 1, 0, 0, 1, 0, 1, 0, 0]  # 1 = resigned
}
df = pd.DataFrame(hr_data)
#Null count per column
print(df.isnull().sum())

#Fill missing Salary with department-wise mean
df["Salary"] = df["Salary"].fillna(df.groupby("Department")["Salary"].transform("mean"))

#Fill missing rating with median
df["Rating"] = df["Rating"].fillna(df["Rating"].median())

#Fill missing department with unknown
df["Department"] = df["Department"].fillna("Unknown")

#Drop rows where name is null
df = df.dropna(subset = "Name")
print(df.to_string(index = False))
```

```python
#Add column "Level" → Experience ≥ 5 = "Senior", else "Junior"
df["Level"] = np.where(df["Experience"] >= 5, "Senior", "Junior")
print(df)
#Average salary per department
print(df.groupby("Department")["Salary"].transform("mean"))
#Count of employees who resigned
resigned_count = df.groupby(df["Left"] == 1).agg(
    count = ("Left", "count")
).reset_index()
print(resigned_count)
#Highest paid employee in each department
print(df.loc[df.groupby("Department")["Salary"].idxmax()][["Department", "Name", "Salary"]])
#Phase 3
print("=====HR Report Summary=====")
print("Total employees:", df["Name"].count())
print("Resigned:", (df["Left"] == 1).sum())
retention = (((df["Left"] == 0).sum())/ (df["Name"].count())) * 100
print(f"Retention rate: {retention:.4f}%")
print(f"Average salary: {df["Salary"].mean():.2f}")
print(f"Highest Salary: {df.loc[df["Salary"].idxmax()]["Name"]}")
print(f"Most Common department: {df["Department"].mode()[0]}")
counts = df["Level"].value_counts()
print(f"Senior employees: {counts["Senior"]}")
print(f"Junior employees: {counts["Junior"]}")
```
