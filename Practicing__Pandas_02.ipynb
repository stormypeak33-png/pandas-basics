Data Cleaning

```python
data = {
    "Name":   ["Alice", "Bob", None, "Diana", "Eve"],
    "Age":    [25, None, 35, 28, 22],
    "Score":  [85, 90, None, 60, 78],
    "City":   ["Chennai", "Mumbai", None, "Pune", "Hyderabad"],
}
df = pd.DataFrame(data)

#Check nulls
print(df.isnull().sum())
print()

#Fill nulls with zero
df["Age"].fillna(0)     #Original df is not changed
df["Age"].fillna(0, inplace = True)       #Original df Changed

#Fill nulls using Average()
df["Score"].fillna(df["Score"].mean(), inplace= True)

#Fill with Text
df["City"].fillna("Unknown", inplace = True)
```

Dropping rows

```python
#Dropping rows where name is null
df.dropna(subset= ["Name"], inplace= True)
print(df)
```

Handling Duplicates

```python
data2 = {
    "Name":  ["Alice", "Bob", "Alice", "Diana"],
    "Score": [85, 90, 85, 60],
}

df2 = pd.DataFrame(data2)
print(df2.duplicated())             #Print true where Duplicates occured

#Dropping duplicates
df2.drop_duplicates(inplace = True)

print(df2)
```

Renaming columns

```python
print("Original df")
print(df)

#Renaming columns
df.rename(columns = {
    "Name": "Student",
    "Score": "Marks"
}, inplace = True)
print()
print("Renamed df")
print(df)                                   #Replacing Name as Student and Score as Mark

#Changing DataType
df["Age"] = df["Age"].astype(int)
```

Aggregations and Groupby


```python
import pandas as pd

data5 = {
    "Name":    ["Ravi", "Priya", "Kiran", "Sneha", "Arjun", "Meena"],
    "City":    ["Chennai", "Mumbai", "Chennai", "Delhi", "Mumbai", "Delhi"],
    "Subject": ["Maths", "Science", "Maths", "English", "Science", "English"],
    "Marks":   [88, 72, 91, 60, 78, 85],
}
df4 = pd.DataFrame(data5)

print(df4["Marks"].sum())    # total
print(df4["Marks"].mean())   # average
print(df4["Marks"].max())    # highest
print(df4["Marks"].min())    # lowest
print(df4["Marks"].count())  # count
```

```python
#Grouping total marks by city
print(df4.groupby("City")["Marks"].sum())

#Average marks by subject
print(df4.groupby("Subject")["Marks"].mean())

#Multiple Aggregation at once
print(df4.groupby("City").agg(
    Total = ("Marks", "sum"),
    Average = ("Marks", "mean"),
    Student_count = ("Name", "count")
))

```

Practice question

```python
#Find total marks of all students
print(df4["Marks"].sum())

#Find average marks per city
print(df4.groupby("City")["Marks"].mean())

#Find the highest marks per subject
print(df4.groupby("Subject")["Marks"].max())

#Count number of students per city
print(df4.groupby("City")["Name"].count())

#Group by city and get the sum and mean of Marks using agg()
print(df4.groupby("City").agg(
    Total = ("Marks", "sum"),
    Average = ("Marks", "mean")
))
```

Merging and Reshaping

```python
students = pd.DataFrame({
    "StudentID": [1, 2, 3, 4],
    "Name":      ["Ravi", "Priya", "Kiran", "Sneha"],
    "DeptID":    [10, 20, 10, 30],
})

departments = pd.DataFrame({
    "DeptID":   [10, 20, 30],
    "DeptName": ["Science", "Arts", "Commerce"],
})

df5 = pd.merge(students, departments, on= "DeptID", how= "left")
print(df5)

#Left , right , etc
```

```python
#Concatenate two dataframes             Joining two DataFrames
batch1 = pd.DataFrame({
    "Name":  ["Arjun", "Meena"],
    "Marks": [88, 75],
})
batch2 = pd.DataFrame({
    "Name":  ["Vijay", "Lakshmi"],
    "Marks": [92, 68],
})

concat = pd.concat([batch1, batch2], ignore_index = True)
print(concat)
```

```python
#Pivot table , reshape data
data6 = {
    "Name":    ["Ravi", "Ravi", "Priya", "Priya"],
    "Subject": ["Maths", "Science", "Maths", "Science"],
    "Marks":   [88, 91, 72, 85],
}
df6 = pd.DataFrame(data6)

pivot = df6.pivot_table(values="Marks", index="Name", columns="Subject")
print(pivot)
```

```python
#Merge types
students = pd.DataFrame({
    "StudentID": [1, 2, 3, 4],
    "Name":      ["Ravi", "Priya", "Kiran", "Sneha"],
    "DeptID":    [10, 20, 10, 30],
})

departments = pd.DataFrame({
    "DeptID":   [10, 20, 30],
    "DeptName": ["Science", "Arts", "Commerce"],
})

df5 = pd.merge(students, departments, on= "DeptID", how= "left")
print(df5)

df6 = pd.merge(students, departments, on= "DeptID", how= "right")
print(df6)

df7 = pd.merge(students, departments, on= "DeptID", how= "inner")
print(df7)

df8 = pd.merge(students, departments, on= "DeptID", how= "outer")
print(df8)
```

```python
from io import StringIO
import pandas as pd

csv_data = """StudentID,Name,Math,Science,English,Attendance,City
1,Ravi,85,90,78,92,Chennai
2,Priya,72,68,80,85,Mumbai
3,Kiran,91,88,95,97,Delhi
4,Sneha,60,55,70,70,Chennai
5,Arjun,78,82,74,88,Hyderabad
6,Meena,55,60,50,65,Mumbai
7,Rohit,88,91,85,93,Delhi
8,Anita,45,50,55,50,Pune
9,Vijay,95,97,92,99,Chennai
10,Lakshmi,68,72,65,75,Hyderabad
11,Suresh,,72,78,60,Pune
12,Kavya,80,85,,88,Mumbai"""

df = pd.read_csv(StringIO(csv_data))

#Inspect the data shape, info and null count
print(df.shape)
print(df.isnull().sum())
print(df.info())

#Fill null values in Math and English with their column mean
df["Math"] = df["Math"].fillna(df["Math"].mean())
df["English"] = df["English"].fillna(df["English"].mean())

#Add a new column "Total" = Math + Science + English
df["Total"] = df["Math"] + df["Science"] + df["English"]

#Add a new column "Average" = Total / 3 (round to 1 decimal)
df["Average"] = (df["Total"]/3).round(1)
print(df)

#Find the top 3 students by Average
print(df.sort_values("Average", ascending = False).head(3))

#Find average Math, Science, English scores per City
print(df.groupby("City").agg(
    Average_Math = ("Math", "mean"),
    Average_science = ("Science", "mean"),
    Average_English = ("English", "mean")
))

#Find students with Average < 65 OR Attendance < 70 (at-risk students)
print(df[(df["Average"]< 65) | (df["Attendance"]< 70)])
```
